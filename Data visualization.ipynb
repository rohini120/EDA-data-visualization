Exploratory data analysis - Data Visualization

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load csv file
df=pd.read_csv(r'C:\Users\guddi\Desktop\ro\titanic_dataset.csv')
```

BARGRAPH

```python
sns.countplot(x='Survived',data=df)
```

```python
sns.countplot(x='Survived',hue='Sex',data=df)
```

LINECHART

```python
sns.lineplot(x='Age',y='Fare',data=df)
```

SCATTERPLOT

```python
df.plot.scatter(x='Age',y='Fare')
```

```python
df.plot.scatter(x='Age',y='Survived')
```

HISTOGRAM

```python
df['Age'].hist(bins=8)
```

```python
df['Age'].hist(bins=80)
```

PIECHART

```python
sizes=df['Survived'].value_counts()
fig1,ax1=plt.subplots()
ax1.pie(sizes,labels=['Not Survived','Survived'],autopct='%1.1f%%',shadow=True)
plt.show()
```
